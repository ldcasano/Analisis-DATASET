```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"/content/bank-full.csv", sep=';', na_values='unknown')

df
```

```python
df.shape
```

```python
df.tail()
```

```python
# miramos cada una de la variables

for column in df.columns:
    print(f'Columna: {column}')
    print(f'Valores nulos: {df[column].isna().sum()}')
    print('------------')
```

```python
# como podemos ver nuestra varible objetivo (poutcome) tiene muchos valores nulos

df = df.dropna(subset=['poutcome'])


# eliminamos los valores nurlos
df.shape
```

```python
for column in df.columns:
    print(f'Columna: {column}')
    print(f'Valores nulos: {df[column].isna().sum()}')
    print('------------')
```

```python
df = df.dropna(subset=['contact'])

for column in df.columns:
    print(f'Columna: {column}')
    print(f'Valores nulos: {df[column].isna().sum()}')
    print('------------')

df.shape
```

```python
df = df.dropna(subset=['education'])
for column in df.columns:
    print(f'Columna: {column}')
    print(f'Valores nulos: {df[column].isna().sum()}')
    print('------------')

df.shape
```

```python
print(df['job'].value_counts())
```

```python
# dumificamos las variables

df.loc[df['poutcome'] == 'success', 'poutcome'] = 1
df.loc[df['poutcome'] == 'failure', 'poutcome'] = 0
df.loc[df['poutcome'] == 'other', 'poutcome'] = 2

df
```

```python
#eliminamos los registros other

df = df[df['poutcome'] != 2]

df.shape

```

```python
df = df.dropna(subset=['job'])
```

```python
print(df['poutcome'].value_counts())

```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
df['poutcome'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Campain New Credits "poutcome"')
plt.xlabel( 'Values')
plt.ylabel('Frecuency')
plt.xticks(ticks=[0, 1], labels=['failure', 'success'], rotation = 360)


plt.show()
```

```python
# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
df['poutcome'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Proporción de Valores en la Variable Dummy')

plt.xticks(ticks=[0, 1], labels=['failure', 'success'])
plt.ylabel('')  # Eliminar la etiqueta del eje y
plt.show()
```
